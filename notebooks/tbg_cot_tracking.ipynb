# TBG-CoT-Bench Tracking Notebook

Run gold trajectories and inspect scenario-level belief dynamics.

```python
from pathlib import Path
import pandas as pd
from tbg_cot_bench.core import load_scenarios, simulate_gold, summarize_trajectory

scenarios = load_scenarios(Path('../scenarios'))
rows = []
summary = []
for s in scenarios:
    traj = simulate_gold(s)
    rows.extend(traj)
    summary.append(summarize_trajectory(s, traj))
pd.DataFrame(summary)
```

```python
import matplotlib.pyplot as plt
df = pd.DataFrame(rows)
for sid, g in df.groupby('scenario_id'):
    plt.figure()
    plt.plot(g['step'], g['p_forward'], marker='o')
    plt.ylim(0, 1)
    plt.title(sid)
    plt.xlabel('Step')
    plt.ylabel('p_forward')
    plt.show()
```
